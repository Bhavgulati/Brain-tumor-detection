```python
import os, torch, torch.nn as nn
from torch.optim import Adam
from torch.utils.data import DataLoader
from torchvision.models import resnet50
from torchvision.transforms import Compose, ToTensor, Resize, Normalize
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
print(f"PyTorch: {torch.__version__}")
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")
```

## Config

```python
TRAIN_DIR = "./Training"
TEST_DIR  = "./Testing"
MODEL_SAVE_PATH = "./models/bt_resnet50_model.pt"
BATCH_SIZE = 32
EPOCHS = 10
LEARNING_RATE = 0.00001
IMAGE_SIZE = 224
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]
CLASS_NAMES = ["Glioma", "Meningioma", "No Tumor", "Pituitary"]
os.makedirs("./models", exist_ok=True)
print("Config ready")
```

## Dataset Exploration

```python
for split, folder in [("Training", TRAIN_DIR), ("Testing", TEST_DIR)]:
    print(f"
{split}:")
    total = 0
    for cls in sorted(os.listdir(folder)):
        p = os.path.join(folder, cls)
        if os.path.isdir(p):
            c = len(os.listdir(p))
            total += c
            print(f"  {cls:<20} {c} images")
    print(f"  TOTAL: {total}")
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,4))
for ax, (split, folder) in zip(axes, [("Training", TRAIN_DIR),("Testing", TEST_DIR)]):
    counts, labels = [], []
    for cls in sorted(os.listdir(folder)):
        p = os.path.join(folder, cls)
        if os.path.isdir(p): counts.append(len(os.listdir(p))); labels.append(cls)
    ax.bar(labels, counts, color=["#fca5a5","#fdba74","#86efac","#a5b4fc"])
    ax.set_title(f"{split} Distribution")
    ax.set_ylabel("Images")
plt.tight_layout(); plt.show()
```

## Data Loading & Transforms
Why 224x224? ResNet50 was designed for this size.
Why ImageNet normalization? The pretrained weights expect this exact pixel range.

```python
transform = Compose([Resize((224,224)), ToTensor(), Normalize(IMAGENET_MEAN, IMAGENET_STD)])
train_dataset = ImageFolder(TRAIN_DIR, transform=transform)
test_dataset  = ImageFolder(TEST_DIR,  transform=transform)
print(f"Classes: {train_dataset.classes}")
print(f"Class->index: {train_dataset.class_to_idx}")
print(f"Train: {len(train_dataset)} | Test: {len(test_dataset)}")
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)
```

## Model Architecture
ResNet50 with custom FC head:
- Linear(2048->2048) + SELU + Dropout(0.4)
- Linear(2048->2048) + SELU + Dropout(0.4)
- Linear(2048->4) — 4 tumor classes

SELU chosen over ReLU: self-normalizing, handles negatives, better for deep FC layers.

```python
model = resnet50(pretrained=True)
for param in model.parameters(): param.requires_grad = True
n = model.fc.in_features
model.fc = nn.Sequential(
    nn.Linear(n, 2048), nn.SELU(), nn.Dropout(0.4),
    nn.Linear(2048, 2048), nn.SELU(), nn.Dropout(0.4),
    nn.Linear(2048, 4)
)
model = model.to(device)
total = sum(p.numel() for p in model.parameters())
print(f"Total params: {total:,}")
print(model.fc)
```

## Training

```python
import time
optimizer = Adam(model.parameters(), lr=LEARNING_RATE)
criterion = nn.CrossEntropyLoss()
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[]}
best_val = 0.0
print(f"{chr(39)*65}")
print(f"Epoch    Train Loss    Train Acc    Val Acc")
print(f"{chr(39)*65}")
for epoch in range(EPOCHS):
    model.train(); rl=0; correct=0; total=0
    for imgs, lbls in train_loader:
        imgs, lbls = imgs.to(device), lbls.to(device)
        optimizer.zero_grad()
        out = model(imgs); loss = criterion(out, lbls)
        loss.backward(); optimizer.step()
        rl += loss.item()
        _, pred = torch.max(out,1); total += lbls.size(0); correct += (pred==lbls).sum().item()
    tacc = 100*correct/total; tloss = rl/len(train_loader)
    model.eval(); vc=0; vt=0; vl=0
    with torch.no_grad():
        for imgs, lbls in test_loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            out = model(imgs); vl += criterion(out,lbls).item()
            _, pred = torch.max(out,1); vt += lbls.size(0); vc += (pred==lbls).sum().item()
    vacc = 100*vc/vt
    history["train_loss"].append(tloss); history["train_acc"].append(tacc)
    history["val_loss"].append(vl/len(test_loader)); history["val_acc"].append(vacc)
    if vacc > best_val: best_val=vacc; torch.save(model.state_dict(), MODEL_SAVE_PATH); saved=" SAVED"
    else: saved=""
    print(f"{epoch+1:<9}{tloss:<14.4f}{tacc:<13.2f}{vacc:.2f}%{saved}")
print(f"Best Val Accuracy: {best_val:.2f}%")
```

## Training Curves

```python
fig, axes = plt.subplots(1,2,figsize=(14,5))
ep = range(1, EPOCHS+1)
axes[0].plot(ep, history["train_loss"], label="Train", color="#f97316")
axes[0].plot(ep, history["val_loss"],   label="Val",   color="#fca5a5", linestyle="--")
axes[0].set_title("Loss"); axes[0].legend(); axes[0].set_xlabel("Epoch")
axes[1].plot(ep, history["train_acc"], label="Train", color="#34d399")
axes[1].plot(ep, history["val_acc"],   label="Val",   color="#86efac", linestyle="--")
axes[1].set_title("Accuracy %"); axes[1].legend(); axes[1].set_xlabel("Epoch")
plt.tight_layout(); plt.savefig("training_curves.png", dpi=150); plt.show()
```

## Evaluation — Confusion Matrix

```python
model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for imgs, lbls in test_loader:
        out = model(imgs.to(device))
        _, pred = torch.max(out,1)
        all_preds.extend(pred.cpu().numpy())
        all_labels.extend(lbls.numpy())
print(classification_report(all_labels, all_preds, target_names=CLASS_NAMES))
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Oranges", xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title("Confusion Matrix"); plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout(); plt.savefig("confusion_matrix.png", dpi=150); plt.show()
```
